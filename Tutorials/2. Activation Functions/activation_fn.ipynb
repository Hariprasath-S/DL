```python
import math
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+math.exp(-z))
```

```python
sigmoid(1)
```

```python
def tanh(z):
    return (math.exp(z) - math.exp(-z)) / (math.exp(z) + math.exp(-z))
```

```python
tanh(1)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-3)
```

```python
relu(5)
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-2)
```

```python
x = list(range(-10,10))
x
```

```python
sigmoid_y = list(map(lambda x: sigmoid(x), x))
plt.plot(x,sigmoid_y)
```

```python
tanh_y = list(map(lambda x: tanh(x), x))
plt.plot(x,tanh_y)
```

```python
relu_y = list(map(lambda x: relu(x), x))
plt.plot(x,relu_y)
```

```python
leaky_relu_y = list(map(lambda x: leaky_relu(x), x))
plt.plot(x,leaky_relu_y)
```
